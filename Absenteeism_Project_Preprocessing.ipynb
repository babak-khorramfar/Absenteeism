# We want to make the Dummy dataset as small as possible:

### Remove first column reason_columns (Multicollinearity)

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first=True)
reason_columns
```

## Remove first column from df (Multicollinearity)

```python
df.columns.values
```

```python
df = df.drop(['Reason for Absence'], axis=1)
df
```

# The reasons for absence are already divided into 4 categories

### So we do the categories

```python
reason_columns.loc[:,1:13]
```

## Combine columns by .max()

```python
reason_type1 = reason_columns.loc[:,1:13].max(axis=1)
reason_type2 = reason_columns.loc[:,14:16].max(axis=1)
reason_type3 = reason_columns.loc[:,17:21].max(axis=1)
reason_type4 = reason_columns.loc[:,22:].max(axis=1)
```

## Test our work by one of them:

```python
reason_type4
```

## Combine df and reason types

```python
df = pd.concat([df, reason_type1, reason_type2, reason_type3, reason_type4], axis=1)
```

```python
df.head()
```

## Change reason's columns names

```python
df.columns.values
```

```python
column_names = ['ID', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average(in Minutes) ', 'Body Mass Index',
       'Education', 'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason1', 'Reason2', 'Reason3', 'Reason4']
```

```python
df.columns = column_names
```

```python
df
```

```python
df = df.drop(['ID'], axis=1)
df
```

```python
reordered_columns = ['Reason1', 'Reason2', 'Reason3', 'Reason4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average(in Minutes) ', 'Body Mass Index',
       'Education', 'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df = df[reordered_columns]
df
```

## Checkpoint

```python
df_reason_modified = df.copy()
```

## Date column type check

```python
type(df_reason_modified['Date'][0])
```

## Convert Date column type from string to Date

```python
df_reason_modified['Date'] = pd.to_datetime(df_reason_modified['Date'])
df_reason_modified['Date'].head()
```

```python
df_reason_modified['Date'].info()
```

## Extracting Month and day

```python
df_reason_modified['Date'][0]
```

```python
list_months = []
```

```python
for i in range(df_reason_modified.shape[0]):
    list_months.append(df_reason_modified['Date'][i].month)
```

```python
df_reason_modified['Month Value'] = list_months
```

```python
weekdays = []
```

```python
for i in range(df_reason_modified.shape[0]):
    weekdays.append(df_reason_modified['Date'][i].weekday())
```

```python
df_reason_modified['Weekday'] = weekdays
```

## Help:

### Week days:
0 => Monday,
1 => Tuesday, 
2 => Wednesday, 
3 => Thursday, 
4 => Friday, 
5 => Saturday, 
6 => Sunday

```python
df_reason_modified.head()
```

## Drop the Date column

```python
df_reason_modified = df_reason_modified.drop(['Date'], axis=1)
df_reason_modified
```

```python
df_reason_modified.columns.values
```

```python
column_names_new= ['Reason1', 'Reason2', 'Reason3', 'Reason4', 'Month Value', 'Weekday', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average(in Minutes) ',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours']
```

```python
df_reason_modified = df_reason_modified[column_names_new]
df_reason_modified.head()
```

## Checkpoint

```python
df_reason_date_modified = df_reason_modified.copy()
```

```python
df_reason_date_modified['Reason1'] = df_reason_date_modified['Reason1'].astype(int)
df_reason_date_modified['Reason2'] = df_reason_date_modified['Reason2'].astype(int)
df_reason_date_modified['Reason3'] = df_reason_date_modified['Reason3'].astype(int)
df_reason_date_modified['Reason4'] = df_reason_date_modified['Reason4'].astype(int)
```

```python
df_reason_date_modified
```
