## Use Median
#### To get the middle of the number of hours that employees are absent, we use the median, which is lower than it is considered as low absence and more than that is considered as high absence and they are represented by 0 and 1 respectively.

```python
data_preprosseced['Absenteeism Time in Hours'].median()
```

```python
target = np.where(data_preprosseced['Absenteeism Time in Hours'] > 3, 1, 0)
target
```

## Create a new column "Excessive Absenteeism"

```python
data_preprosseced['Excessive Absenteeism'] = target
data_preprosseced.head()
```
